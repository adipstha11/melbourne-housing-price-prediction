### Problem 2.1: Preprocessing for the Baseline Models

Let's start by preprocessing the data so that we can train some baseline models. Note that categorical variables will need to be encoded and missing values should be appropriately delt with using imputation. Also, some models may require scaling. Imputation and scaling should be done in a manner that aovids data snooping.
> **Hint:** The use of Scikit-Learn's Pipelines are highly encouraged!

```python
# your code here
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Note: here we are separating input (X) and output (y)
X = data.drop(columns=['Price'])
y = data['Price']

# Note: Now we would like to indentfy the numerical and categorial columns
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

#Note: We would no do numerical preprocessing whih is imputing the missing values and scaling it. 
numerical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')),('scaler', StandardScaler())])

# Note: here we are doing the categorical preprocessing which is to impute the missig values and using encoder; one hot encode 
categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),('encoder', OneHotEncoder(handle_unknown='ignore'))])

# Note: Now Combining transformations
preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_features),('cat', categorical_transformer, categorical_features)])

# Note: We would like to fit only on the training data which is to avoid data snooping or leakage
preprocessor.fit(X_train)
```

### Problem 2.2: Perform Linear Regression

Perform Linear Regression on training data. Predict the output for the test dataset using the fitted model. Print the root mean squared error (RMSE) and R^2 value from Linear Regression for test data set.


```python
# your code here
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Note: We are creating a full pipe line which is connecting preprocessing and the model
linear_model = Pipeline(steps=[('preprocessor', preprocessor),('model', LinearRegression())])

#Note: next step would be to train the model on our training data 
linear_model.fit(X_train, y_train)

# Note: predicting on the test set 
y_pred = linear_model.predict(X_test)

# Note: we would use that prediction to Compute metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Linear Regression Results:")
print(f"RMSE: {rmse:.4f}")
print(f"R^2: {r2:.4f}")

```

### Problem 2.3 : Perform K-Nearest Neighbors Regression

Perform K-Nearest Neighbors Algorithm for Regression using cross validation to tune any hyperparameters, such as `n_neighbors`.

Then, using the optimal hyperparameters, train a K-Nearest Neighbors Regressor on the entirety of the training data and predict output for the test dataset using the fitted model.

Print root mean squared error (RMSE) and R^2 from K-Nearest Neighbors Regression on the test datset.

```python
# your code here
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Note : Step 1 here would be to again as above build  a full pipeline from preprocessing to model
knn_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', KNeighborsRegressor())])

# Note: next step is to tune n_neighbour using crossvalidation
n_neighbours = {'model__n_neighbors': [3, 5, 7, 9, 11, 15, 20]}

knn_grid_for_cv = GridSearchCV(knn_pipeline,n_neighbours,cv=kfolds,scoring="neg_root_mean_squared_error",n_jobs=-1)

knn_grid_for_cv.fit(X_train, y_train)

print("The best KNN hyperparameters:", knn_grid_for_cv.best_params_)

# Note: finally training the final model and evaluating it on our test set
y_pred = knn_grid_for_cv.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("KNN Regression Results:")
print("RMSE:", rmse)
print("R^2:", r2)

```

### Problem 2.4 : Perform Decision Tree Regression 

Perform Decision Tree Regression using cross validation to tune any hyperparameters, such as `max_depth`.

Then, using the optimal hyperparameters, train a Decision Tree Regressor on the entirety of the training data and predict output for the test dataset using the fitted model.

Print root mean squared error (RMSE) and R^2 from Decision Tree Regression on the test dataset.

```python
# Note: We are importing the Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

# Note: Here we would build a pipeline that includes from preprocessing to the decision tree model 
dec_tree_pipeline = Pipeline(steps=[ ('preprocessor', preprocessor), ('model', DecisionTreeRegressor(random_state=42))])

# Note: we are building a hyperparameter grid for runing now. Where - 
# max_depth controls how deep the tree can grow
# min_samples_split controls how many samples needed to split a node
# min_samples_leaf controls how many samples minimum per leaf node
dec_tree_hyperparam = {'model__max_depth': [3, 5, 7, 10, 15, 20, None],'model__min_samples_split': [2, 5, 10],'model__min_samples_leaf': [1, 2, 4]}

# Note: Now we do GridSearchCV which performs cross-validation on each combination
dec_tree_grid = GridSearchCV( dec_tree_pipeline,dec_tree_hyperparam, cv=kfolds,scoring="neg_root_mean_squared_error", n_jobs=-1)

# Note: Next again as needed Fitting the model on the training data and we use CV to find the best parameters
dec_tree_grid.fit(X_train, y_train)

# Note: We are just printing the best hyperparameters found
print("Best Decision Tree hyperparameters:", dec_tree_grid.best_params_)

# Note: Finally Predicting on the test set using the best model trained on the training data
y_pred = dec_tree_grid.predict(X_test)

# Note: using the prediction to compute evaluation metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


print("Decision Tree Regression Results:")
print("RMSE:", rmse)
print("R^2:", r2)

```

Which makes predictions that can fall apart when face with new information. Random Forests, on the other hand, are way smarter. Instead of relying on one tree's opinion, they build an entire forest of trees(multiple trees) each tree is trained in a slightly different random sample of the data. This is why its so good all the different trees are evaluated and the averages their predictions  which leads to cancellation of each individual trees mistakes. The overfitting in single trees are wiped out and produces a more smoother model. Its also not just about building different various trees but random forest uses this technique called bagging or called bootstrap aggregating where each tree is trained on random subset of the data, which means  not two trees will be completely the same which leeds to a more stable average. A single decision tree is prone to high variance. Also at each split random forrest only considers random subset of features instead of all features which helps prevent single strong predictor and also encourages tree diversity.

```python
# your code here

# Note: first we make copies of the original X_train and X_test to avoid messing with the origincal data
X_train_for_feature_ex = X_train.copy()
X_test_for_feature_ex = X_test.copy()


# Note : we would like to use Feature 1 as  HouseAge
# The Reason are as follows :
#    YearBuilt has many missing values and are not directly usefull here
#    HouseAge = 2018 - YearBuilt (since dataset ends in 2018).
#    Older homes may be cheaper or renovated, adding predictive signal.


# Note: Now we fill missing YearBuilt values with our median just inthe training set\
yearbuilt_median = X_train_for_feature_ex['YearBuilt'].median()

X_train_for_feature_ex['HouseAge'] = 2018 - X_train_for_feature_ex['YearBuilt'].fillna(yearbuilt_median)
X_test_for_feature_ex ['HouseAge'] = 2018 - X_test_for_feature_ex ['YearBuilt'].fillna(yearbuilt_median)



# Note: Feature 2 - BathroomPerRoom
# The reasons are as follows:
#   Homes with very few bathrooms relative to their number of rooms may be less desirable.
#   This ratio captures bathroom availability per room, a useful measure of livability.
#   It also converts two features into one meaningful interaction feature.

# Note: Avoid division by zero by replacing Rooms == 0 with 1
X_train_for_feature_ex['BathroomPerRoom'] = X_train_for_feature_ex['Bathroom'] / X_train_for_feature_ex['Rooms'].replace(0, 1)
X_test_for_feature_ex['BathroomPerRoom'] = X_test_for_feature_ex['Bathroom'] / X_test_for_feature_ex['Rooms'].replace(0, 1)

# Print changes
print("New features added: 'HouseAge' and 'BathroomPerRoom'")
print(X_train_for_feature_ex[['HouseAge', 'BathroomPerRoom']].head())

```
